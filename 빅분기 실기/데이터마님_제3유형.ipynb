```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/anova.csv')

df

from scipy import stats
print(stats.shapiro(df.x1))
print(stats.shapiro(df.x2))
print(stats.shapiro(df.x3))

print(stats.levene(df.x1, df.x2, df.x3)) # 정규성, 등분산성 만족

from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

df_m = df.melt()

model = ols('value ~ variable', data = df_m).fit()
print(anova_lm(model))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/anova2.csv')

from scipy import stats
print(stats.shapiro(df.x1))
print(stats.shapiro(df.x2))
print(stats.shapiro(df.x3))

print(stats.levene(df.x1, df.x2, df.x3))

print(stats.f_oneway(df.x1, df.x2, df.x3))

from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

dfm = df.melt()
model = ols("value ~ variable", data =dfm).fit()
print(anova_lm(model))
```
