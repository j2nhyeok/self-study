```python
# ÏßàÎ¨∏ 3
df3 = df[['channelTitle', 'channelId']]
df3 = df3.drop_duplicates(['channelTitle', 'channelId'], keep = 'first')
df4 = df3.groupby(['channelId']).size().reset_index()
df4

cond = df4[0] > 1
print(len(df4[cond]))
```

```python
# ÏßàÎ¨∏ 4
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
df['weekday'] = df['trending_date2'].dt.dayofweek

cond = df['weekday'] == 6
df5 = df[cond]
df5.groupby('categoryId').size().sort_values(ascending=False).index[0]
```

```python
# ÏßàÎ¨∏ 5 Í∞Å ÏöîÏùºÎ≥Ñ Ïù∏Í∏∞ ÏòÅÏÉÅÎì§Ïùò categoryIdÎäî Í∞ÅÍ∞Å Î™áÍ∞ú Ïî©Ïù∏ÏßÄ ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏúºÎ°ú ÌëúÌòÑÌïòÎùº
group = df.groupby([df['trending_date2'].dt.day_name(), df['categoryId']], as_index=False).size()
answer = group.pivot(index = 'categoryId', columns = 'trending_date2')
answer
```

```python
# ÏßàÎ¨∏ 6 ÎåìÍ∏ÄÏùò ÏàòÎ°ú (comment_count) ÏòÅÏÉÅ Î∞òÏùëÏóê ÎåÄÌïú ÌåêÎã®ÏùÑ Ìï† Ïàò ÏûàÎã§. viewcountÎåÄÎπÑ ÎåìÍ∏ÄÏàòÍ∞Ä Í∞ÄÏû• ÎÜíÏùÄ ÏòÅÏÉÅÏùÑ ÌôïÏù∏ÌïòÎùº (view_countÍ∞íÏù¥ 0Ïù∏ Í≤ΩÏö∞Îäî Ï†úÏô∏ÌïúÎã§)

cond = df['view_count'] != 0
df2 = df[cond]
df2['ratio'] = df2['comment_count'] / df2['view_count']
df2.sort_values(by = 'ratio', ascending = False).iloc[0, 0]
```

```python
# ÏßàÎ¨∏ 7 ÎåìÍ∏ÄÏùò ÏàòÎ°ú (comment_count) ÏòÅÏÉÅ Î∞òÏùëÏóê ÎåÄÌïú ÌåêÎã®ÏùÑ Ìï† Ïàò ÏûàÎã§.viewcountÎåÄÎπÑ ÎåìÍ∏ÄÏàòÍ∞Ä Í∞ÄÏû• ÎÇÆÏùÄ ÏòÅÏÉÅÏùÑ ÌôïÏù∏ÌïòÎùº (view_counts, ratioÍ∞íÏù¥ 0Ïù∏Í≤ΩÏö∞Îäî Ï†úÏô∏ÌïúÎã§.)
df2 = df2.sort_values(by = 'ratio')
cond = df2['ratio'] != 0
df2 = df2[cond]
df2.iloc[0, 0]
```

```python
# ÏßàÎ¨∏ 8 ike ÎåÄÎπÑ dislikeÏùò ÏàòÍ∞Ä Í∞ÄÏû• Ï†ÅÏùÄ ÏòÅÏÉÅÏùÄ Î¨¥ÏóáÏù∏Í∞Ä? (like, dislike Í∞íÏù¥ 0Ïù∏Í≤ΩÏö∞Îäî Ï†úÏô∏ÌïúÎã§)
cond = (df['likes'] != 0) & (df['dislikes'] != 0)
df2 = df[cond]
df2['ratio'] = df2['dislikes'] / df2['likes']
df2 = df2.sort_values('ratio', ascending=True)
df2.iloc[0,0]
```

```python
# ÏßàÎ¨∏ 9 Í∞ÄÏû• ÎßéÏùÄ Ìä∏Î†åÎìú ÏòÅÏÉÅÏùÑ Ï†úÏûëÌïú Ï±ÑÎÑêÏùò Ïù¥Î¶ÑÏùÄ Î¨¥ÏóáÏù∏Í∞Ä? (ÎÇ†Ïßú Í∏∞Ï§Ä, Ï§ëÎ≥µ Ìè¨Ìï®)
df.groupby(['channelTitle']).trending_date2.size().sort_values(ascending =False).index[0]
```

```python
# ÏßàÎ¨∏ 10 20Ìöå(20Ïùº)Ïù¥ÏÉÅ Ïù∏Í∏∞ÎèôÏòÅÏÉÅ Î¶¨Ïä§Ìä∏Ïóê Ìè¨Ìï®Îêú ÎèôÏòÅÏÉÅÏùò Ïà´ÏûêÎäî?

df2 = (df[['title', 'channelId']].value_counts() >= 20).sum()
df2
```
