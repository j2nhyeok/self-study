# 유튜브 인기 동영상 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()
```

```python
# 질문 1
list(df.groupby(['channelTitle']).size().sort_values(ascending=False).head(10).index)
```

```python
# 질문 2
cond = df['dislikes'] > df['likes']
df2 = df[cond]
list(df2['channelTitle'].unique())
```

```python
# 질문 3
df3 = df[['channelTitle', 'channelId']]
df3 = df3.drop_duplicates(['channelTitle', 'channelId'], keep = 'first')
df4 = df3.groupby(['channelId']).size().reset_index()
df4

cond = df4[0] > 1
print(len(df4[cond]))
```

```python
# 질문 4
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
df['weekday'] = df['trending_date2'].dt.dayofweek

cond = df['weekday'] == 6
df5 = df[cond]
df5.groupby('categoryId').size().sort_values(ascending=False).index[0]
```

```python
# 질문 5 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라
group = df.groupby([df['trending_date2'].dt.day_name(), df['categoryId']], as_index=False).size()
answer = group.pivot(index = 'categoryId', columns = 'trending_date2')
answer
```

```python
# 질문 6 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다. viewcount대비 댓글수가 가장 높은 영상을 확인하라 (view_count값이 0인 경우는 제외한다)

cond = df['view_count'] != 0
df2 = df[cond]
df2['ratio'] = df2['comment_count'] / df2['view_count']
df2.sort_values(by = 'ratio', ascending = False).iloc[0, 0]
```

```python
# 질문 7 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)
df2 = df2.sort_values(by = 'ratio')
cond = df2['ratio'] != 0
df2 = df2[cond]
df2.iloc[0, 0]
```

```python
# 질문 8 ike 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)
cond = (df['likes'] != 0) & (df['dislikes'] != 0)
df2 = df[cond]
df2['ratio'] = df2['dislikes'] / df2['likes']
df2 = df2.sort_values('ratio', ascending=True)
df2.iloc[0,0]
```

```python
# 질문 9 가장 많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? (날짜 기준, 중복 포함)
df.groupby(['channelTitle']).trending_date2.size().sort_values(ascending =False).index[0]
```

```python
# 질문 10 20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?

df2 = (df[['title', 'channelId']].value_counts() >= 20).sum()
df2
```

# 유튜브 공범컨텐츠 동영상 데이터

```python
import pandas as pd


channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

```python
# 질문 11
video['ct'] = pd.to_datetime(video['ct'])
video['videoname'].value_counts()
```

```python
# 질문 12 수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라
video.sort_values(['videoname', 'ct']).drop_duplicates('videoname', keep='last')[['viewcnt', 'videoname', 'ct']].reset_index(drop=True)
```

```python
#질문 13
channel['ct'] = pd.to_datetime(channel['ct'])

cond = channel['ct'] >= '2021-10-03'
channel2 = channel[cond]
channel2 = channel2.drop_duplicates('channelname', keep='first')[['channelname', 'subcnt']].reset_index(drop=True)
channel2
```

```python
# 질문 14
end = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-11-01 15']
start = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-10-03 03']

end_df = end[['channelname', 'subcnt']].reset_index(drop=True)
start_df = start[['channelname', 'subcnt']].reset_index(drop =True)

end_df.columns = ['channelname', 'end_sub']
start_df.columns = ['channelname', 'start_sub']

tt = pd.merge(start_df, end_df)
tt['del'] = tt['end_sub'] - tt['start_sub']
result = tt[['channelname', 'del']]
display(result)
```

```python
# 질문 16
video['date'] = video['ct'].dt.date
video.sort_values(['videoname', 'date']).drop_duplicates('videoname', keep='first')[['date', 'videoname']]
```

```python
# 질문 17
video['hour'] = video['ct'].dt.hour

cond = video['hour'] == 21
video2 = video[cond]
video2.drop_duplicates('videoname')[['videoname', 'viewcnt', 'ct']].sort_values('viewcnt', ascending= False)
```

```python
# 질문 18
df = video.drop_duplicates('videoname', keep = 'last')
df['ratio'] = df['dislikecnt'] / df['likecnt']
df[['videoname', 'ratio']].sort_values('ratio')
```
