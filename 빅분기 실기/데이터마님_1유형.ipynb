```python
# 질문 1
list(df.groupby(['channelTitle']).size().sort_values(ascending=False).head(10).index)
```

```python
# 질문 3
df3 = df[['channelTitle', 'channelId']]
df3 = df3.drop_duplicates(['channelTitle', 'channelId'], keep = 'first')
df4 = df3.groupby(['channelId']).size().reset_index()
df4

cond = df4[0] > 1
print(len(df4[cond]))
```

```python
# 질문 4
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
df['weekday'] = df['trending_date2'].dt.dayofweek

cond = df['weekday'] == 6
df5 = df[cond]
df5.groupby('categoryId').size().sort_values(ascending=False).index[0]
```

```python
# 질문 5 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라
group = df.groupby([df['trending_date2'].dt.day_name(), df['categoryId']], as_index=False).size()
answer = group.pivot(index = 'categoryId', columns = 'trending_date2')
answer
```

```python
# 질문 6 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다. viewcount대비 댓글수가 가장 높은 영상을 확인하라 (view_count값이 0인 경우는 제외한다)

cond = df['view_count'] != 0
df2 = df[cond]
df2['ratio'] = df2['comment_count'] / df2['view_count']
df2.sort_values(by = 'ratio', ascending = False).iloc[0, 0]
```

```python
# 질문 7 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)
df2 = df2.sort_values(by = 'ratio')
cond = df2['ratio'] != 0
df2 = df2[cond]
df2.iloc[0, 0]
```

```python
# 질문 8 ike 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)
cond = (df['likes'] != 0) & (df['dislikes'] != 0)
df2 = df[cond]
df2['ratio'] = df2['dislikes'] / df2['likes']
df2 = df2.sort_values('ratio', ascending=True)
df2.iloc[0,0]
```

```python
# 질문 9 가장 많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? (날짜 기준, 중복 포함)
df.groupby(['channelTitle']).trending_date2.size().sort_values(ascending =False).index[0]
```

```python
# 질문 10 20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?

df2 = (df[['title', 'channelId']].value_counts() >= 20).sum()
df2
```

# 의료 비용 예측 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
df.head()
```

```python
df_smoker = df[df['smoker'] == "yes"]
df_non = df[df['smoker'] == "no"]

smoker = df_smoker['charges'].quantile(0.9)
non = df_non['charges'].quantile(0.9)

cond1 = df_smoker['charges'] >= smoker
cond2 = df_non['charges'] >= non

a = df_smoker[cond1].charges.mean()
b = df_non[cond2].charges.mean()

print(b - a)
```
