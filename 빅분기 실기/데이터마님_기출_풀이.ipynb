```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_test.csv")

# EDA
#print(x_train.shape, y_train.shape, x_test.shape)  # ID값 버리기
#x_train.info() # 널값 없음 encoding 필요
# x_train.describe()  #'Weight_in_gms', 'Cost_of_the_Product', 'Discount_offered'  스케일링 필요

# 전처리 (기타)
x_train = x_train.drop('ID', axis = 1)
x_test = x_test.drop('ID', axis = 1)
y_train = y_train.drop('ID', axis = 1)

cols_obj = list(x_train.select_dtypes(include = 'object').columns)
cols_num = list(x_train.select_dtypes(exclude = 'object').columns)

scale_list = ['Weight_in_gms', 'Cost_of_the_Product', 'Discount_offered']
# 전처리 (인코딩)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cols_obj:
  x_train[col] = le.fit_transform(x_train[col])
  x_test[col] = le.transform(x_test[col])

# 전처리 (스케일링)
from sklearn.preprocessing import RobustScaler

rs = RobustScaler()
x_train[scale_list] = rs.fit_transform(x_train[scale_list])
x_test[scale_list] = rs.transform(x_test[scale_list])

# 검증 데이터 분리
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(x_train, y_train, test_size = 0.15, random_state = 2024)

# 모델링 학습
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state = 2024, max_depth = 3, n_estimators = 400)
clf.fit(X_tr, y_tr)
pred = clf.predict(X_val)

# 평가
from sklearn.metrics import f1_score
print(f1_score(y_val, pred))

# 예측
pred = clf.predict(x_test)
submit = pd.DataFrame({
    'pred' : pred
}).to_csv('sdff.csv', index = False)
```

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

X_train = train.drop('Segmentation', axis = 1)
y_train = train['Segmentation']

# EDA
# print(X_train.info())  # ID 값 제거, 인코딩 필요, 널값 많음
# print(X_train.describe(include = 'O'))
# print(X_train.isnull().sum())
# 전처리 (기타)
MAX_EVER_MARRIED = train['Ever_Married'].value_counts().index[0]
MAX_GRADUATED = train['Graduated'].value_counts().index[0]
MAX_PROFESSION = train['Profession'].value_counts().index[0]
MEAD_Work_Experience = train['Work_Experience'].mean()
MAX_Family_Size = train['Family_Size'].value_counts().index[0]
MAX_Var_1 = train['Var_1'].value_counts().index[0]

fillna_dic = {
    'Ever_Married':MAX_EVER_MARRIED,
    'Graduated':MAX_GRADUATED,
    'Profession':MAX_PROFESSION,
    'Work_Experience':MEAD_Work_Experience,
    'Family_Size':MAX_Family_Size,
    'Var_1':MAX_Var_1
}

for col in fillna_dic.keys():
    X_train.loc[:,col] = X_train.loc[:,col].fillna(fillna_dic[col])
    test.loc[:,col] = test.loc[:,col].fillna(fillna_dic[col])

X_train = X_train.drop('ID', axis = 1)
test = test.drop('ID', axis = 1)

cols_obj = list(X_train.select_dtypes(include='object').columns)
cols_num = list(X_train.select_dtypes(exclude='object').columns)

# 전처리 (인코딩)
from sklearn.preprocessing import LabelEncoder

for col in cols_obj:
  le = LabelEncoder()
  X_train[col] = le.fit_transform(X_train[col])
  test[col] = le.transform(test[col])

# 전처리 (스케일링)

# 검증 데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.15, random_state = 2022, stratify=y_train)

# 모델링 학습
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state = 2024, max_depth = 6, n_estimators = 100)
clf.fit(X_tr, y_tr)
pred = clf.predict(X_val)


# 평가 (f1-score macro)
from sklearn.metrics import f1_score
print(f1_score(y_val, pred, average = 'macro'))
# 예측
pred = clf.predict(test)
submit = pd.DataFrame({
    "dsfs" : pred
}).to_csv('dsfsf.csv', index = False)
```
