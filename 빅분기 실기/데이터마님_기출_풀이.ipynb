```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

X_train = train.drop('Segmentation', axis = 1)
y_train = train['Segmentation']

# EDA
# print(X_train.info())  # ID 값 제거, 인코딩 필요, 널값 많음
# print(X_train.describe(include = 'O'))
# print(X_train.isnull().sum())
# 전처리 (기타)
MAX_EVER_MARRIED = train['Ever_Married'].value_counts().index[0]
MAX_GRADUATED = train['Graduated'].value_counts().index[0]
MAX_PROFESSION = train['Profession'].value_counts().index[0]
MEAD_Work_Experience = train['Work_Experience'].mean()
MAX_Family_Size = train['Family_Size'].value_counts().index[0]
MAX_Var_1 = train['Var_1'].value_counts().index[0]

fillna_dic = {
    'Ever_Married':MAX_EVER_MARRIED,
    'Graduated':MAX_GRADUATED,
    'Profession':MAX_PROFESSION,
    'Work_Experience':MEAD_Work_Experience,
    'Family_Size':MAX_Family_Size,
    'Var_1':MAX_Var_1
}

for col in fillna_dic.keys():
    X_train.loc[:,col] = X_train.loc[:,col].fillna(fillna_dic[col])
    test.loc[:,col] = test.loc[:,col].fillna(fillna_dic[col])

X_train = X_train.drop('ID', axis = 1)
test = test.drop('ID', axis = 1)

cols_obj = list(X_train.select_dtypes(include='object').columns)
cols_num = list(X_train.select_dtypes(exclude='object').columns)

# 전처리 (인코딩)
from sklearn.preprocessing import LabelEncoder

for col in cols_obj:
  le = LabelEncoder()
  X_train[col] = le.fit_transform(X_train[col])
  test[col] = le.transform(test[col])

# 전처리 (스케일링)

# 검증 데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.15, random_state = 2022, stratify=y_train)

# 모델링 학습
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state = 2024, max_depth = 6, n_estimators = 100)
clf.fit(X_tr, y_tr)
pred = clf.predict(X_val)


# 평가 (f1-score macro)
from sklearn.metrics import f1_score
print(f1_score(y_val, pred, average = 'macro'))
# 예측
pred = clf.predict(test)
submit = pd.DataFrame({
    "dsfs" : pred
}).to_csv('dsfsf.csv', index = False)
```

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

train = train.drop('ID', axis = 1)
test = test.drop('ID', axis = 1)

X_train = train.drop('이용금액', axis = 1)
y_train = train['이용금액']
X_test = test.copy()

# EDA
# print(X_train.shape, y_train.shape, X_test.shape)
# print(X_train.info())
# print(X_train.describe()) # ['이용자수', '이용건수']



# 전처리(기타)
X_train['연월'] = X_train['연월'].astype(str)
X_test['연월'] = X_test['연월'].astype(str)

cols_obj = list(X_train.select_dtypes(include = 'object').columns)
cols_num = list(X_train.select_dtypes(exclude = 'object').columns)

cols_scale = ['이용자수', '이용건수']

# 전처리(인코딩)
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()

# for col in cols_obj:
#   X_train[col] = le.fit_transform(X_train[col])
#   X_test[col] = le.transform(X_test[col])


X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
X_test = X_test.reindex(columns = X_train.columns, fill_value = 0)



# 전처리(스케일링)

# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.15, random_state = 2022)

# 학습
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state = 2022, max_depth = 12, n_estimators = 300)
model.fit(X_tr, y_tr)
pred = model.predict(X_val)

# 평가
from sklearn.metrics import mean_squared_error
import numpy as np

def rmse(y_val, pred):
  return np.sqrt(mean_squared_error(y_val, pred))

print(rmse(y_val, pred))

# 예측
pred = model.predict(X_test)
submit = pd.DataFrame({
    '이용금액':pred
}).to_csv('asdfs.csv' ,index=False)
```
