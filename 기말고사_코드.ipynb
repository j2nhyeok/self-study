```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/dsf/5대+범죄+발생현황_20240605144456.csv', index_col='자치구',  encoding='euc-kr')

# 숫자가 문자열로 저장되어 있어서 숫자 형식으로 변환
columns_to_convert = [
    '2018_소계', '2019_소계', '2020_소계', '2021_소계', '2022_소계',
    '2018_살인', '2019_살인', '2020_살인', '2021_살인', '2022_살인',
    '2018_강도', '2019_강도', '2020_강도', '2021_강도', '2022_강도',
    '2018_강간/강제추행', '2019_강간/강제추행', '2020_강간/강제추행', '2021_강간/강제추행', '2022_강간/강제추행',
    '2018_절도', '2019_절도', '2020_절도', '2021_절도', '2022_절도',
    '2018_폭력', '2019_폭력', '2020_폭력', '2021_폭력', '2022_폭력'
]

for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 각 자치구별 최근 5년간 데이터 합계 계산
columns_to_sum = {
    '소계': ['2018_소계', '2019_소계', '2020_소계', '2021_소계', '2022_소계'],
    '살인': ['2018_살인', '2019_살인', '2020_살인', '2021_살인', '2022_살인'],
    '강도': ['2018_강도', '2019_강도', '2020_강도', '2021_강도', '2022_강도'],
    '강간/강제추행': ['2018_강간/강제추행', '2019_강간/강제추행', '2020_강간/강제추행', '2021_강간/강제추행', '2022_강간/강제추행'],
    '절도': ['2018_절도', '2019_절도', '2020_절도', '2021_절도', '2022_절도'],
    '폭력': ['2018_폭력', '2019_폭력', '2020_폭력', '2021_폭력', '2022_폭력']
}

for new_col, cols in columns_to_sum.items():
    df[f'최근 5년간 {new_col}'] = df[cols].sum(axis=1)

# '최근 5년간 소계' 열로 내림차순 정렬
df = df.sort_values('최근 5년간 소계', ascending= False)
df.to_csv('5대+범죄+발생현황_20240605144456.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
file_path = '/content/drive/MyDrive/5대+범죄+발생현황_20240605144456.csv'
data = pd.read_csv(file_path, encoding='cp949')

# 송파와 중구의 데이터를 추출
songpa = data[data['자치구'] == '송파구']
junggu = data[data['자치구'] == '중구']

# 서울 자치구 평균 인구 수 계산
seoul_mean_population = data['인구'].mean()

# 가설 1: 송파의 인구 수가 서울 자치구 평균 인구 수보다 높은지 확인
songpa_population = songpa['인구'].values[0]
is_songpa_population_higher = songpa_population > seoul_mean_population

# 서울 자치구 평균 범죄 발생 건수 계산
seoul_mean_crime = data['최근 5년간 소계'].mean()

# 가설 1: 송파의 범죄 발생 건수가 서울 자치구 평균보다 높은지 확인
songpa_crime = songpa['최근 5년간 소계'].values[0]
is_songpa_crime_higher = songpa_crime > seoul_mean_crime

# 가설 2: 송파구와 중구의 인구 대비 범죄 발생률 비교
songpa_crime_rate = songpa['최근 5년간 소계'].values[0] / songpa['인구'].values[0] * 100
junggu_crime_rate = junggu['최근 5년간 소계'].values[0] / junggu['인구'].values[0] * 100
is_songpa_crime_rate_lower = songpa_crime_rate < junggu_crime_rate

# 결과 출력
print(f"서울 자치구 평균 인구 수: {seoul_mean_population}")
print(f"송파구 인구 수: {songpa_population}")
print(f"송파구의 인구 수가 서울 자치구 평균 인구 수보다 높습니다: {is_songpa_population_higher}")

print(f"서울 자치구 평균 범죄 발생 건수: {seoul_mean_crime}")
print(f"송파구 범죄 발생 건수: {songpa_crime}")
print(f"송파구의 범죄 발생 건수가 서울 자치구 평균보다 높습니다: {is_songpa_crime_higher}")

print(f"송파구 인구 대비 범죄 발생률: {songpa_crime_rate}%")
print(f"중구 인구 대비 범죄 발생률: {junggu_crime_rate}%")
print(f"송파구의 인구 대비 범죄 발생률이 중구보다 낮습니다: {is_songpa_crime_rate_lower}")

# 시각화 (옵션)
# Plot the population and crime rates for each district
plt.figure(figsize=(14, 8))
data.plot(kind='bar', x='자치구', y=['최근 5년간 소계'], ax=plt.gca(), color='skyblue')
plt.title('자치구별 최근 5년간 범죄 발생 건수')
plt.xlabel('자치구')
plt.ylabel('범죄 발생 건수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the crime rate per capita
data['인구 대비 범죄율'] = data['최근 5년간 소계'] / data['인구'] * 100
plt.figure(figsize=(14, 8))
data.plot(kind='bar', x='자치구', y='인구 대비 범죄율', ax=plt.gca(), color='skyblue')
plt.title('자치구별 인구 대비 범죄율')
plt.xlabel('자치구')
plt.ylabel('인구 대비 범죄율 (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
