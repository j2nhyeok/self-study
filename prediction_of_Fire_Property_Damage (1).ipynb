```python
spots = ['강남구', '마포구', '강서구', '강동구', '양천구', '서초구', '송파구', '관악구', '서대문구',
       '은평구', '영등포구', '중랑구', '노원구', '광진구', '강북구', '성북구', '성동구', '종로구',
       '용산구', '중구', '동작구', '동대문구', '금천구', '구로구', '도봉구']
years = [2017, 2018, 2019, 2020, 2021]
```

```python
for spot in spots:
  for year in years:
    cond1 = data['시군구명'] == spot
    cond2 = data['화재발생연도'] == year

    data.loc[cond1 & cond2, '소방서 1개소당 담당인구(명)'] = FS_in_charge.loc[spot, f"{year}-소방서 1개소당 담당인구 (명)"]
    data.loc[cond1 & cond2, '소방공무원 1인당 담당인구(명)'] = FS_in_charge.loc[spot, f"{year}-소방공무원1인당담당인구 (명)"]

```
