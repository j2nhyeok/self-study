## 문제 1. 
- 다음은 California Housing 데이터 세트이다. 데이터 중 결측치가 있는 경우 해당 데이터의 행을 모두 제거하고, 첫번째 행부터 순서대로 70%까지의 데이터를 훈련 데이터로 추출한 데이터 세트를 구성한다. 변수 중 'housing_media_age'의 Q1(제 1사분위수) 값을 정수로 계산하시오. 

```python
import pandas as pd 
import numpy as np

df = pd.read_csv("data/3회/301_housing.csv")
df.head()
```

```python
## 결측치가 있는 행 모두 제거
df_dropna = df.dropna().reset_index()

print("원본데이터 건수: ", len(df))
print("결측치 제거 후 데이터 건수: ", len(df_dropna))
```

```python
# 첫번째 행부터 순서대로 70%까지의 데이터를 훈련 데이터로 추출한 데이터 세트를 구성한다.
train_df = df_dropna.iloc[0:int(len(df_dropna) * 0.7), :]

print("훈련 데이터 70% 데이터 건수:", len(train_df))
```

```python
# 변수 중 'housing_media_age'의 Q1(제 1사분위수) 값을 정수로 계산하시오.
print(int(np.quantile(train_df['housing_median_age'], 0.25)))
print(int(train_df['housing_median_age'].quantile(0.25)))
```

## 문제 2. 
- 다음은 국가별 연도별 인구 10만 명당 결핵 유병률 데이터 세트이다. 2000년도의 국가별 결핵 유병률 데이터 세트에서 2000년도의 평균값보다 더 큰 유병률 값을 가진 국가의 수를 계산하시오. 

```python
import numpy as np
import pandas as pd 

df = pd.read_csv("data/3회/302_worlddata.csv")
df.head()
```

```python
countries = list(df.columns)
countries.remove("year")
countries[0:3]
```

```python
df_2000 = df[df['year'] == 2000].reset_index(drop=True)
df_2000_T = df_2000.melt(id_vars = 'year', value_vars = countries, value_name = "유병률", var_name = "country")
rst_df = df_2000_T[df_2000_T['유병률'] > df_2000_T['유병률'].mean()]
rst_df
```

### 방법 2

```python
df_2000 = df[df['year'] == 2000].drop('year', axis=1)
df_2000.index = ['value']
df_2000_T = df_2000.T
print(df_2000_T)
```

```python
rst_df = df_2000_T[df_2000_T['value'] > df_2000_T['value'].mean()]
rst_df
```

## 문제 3
- 다음은 Titanic 데이터 세트이다. 주어진 데이터 세트의 컬럼 중 빈 값 또는 결측치를 확인하여 결측치의 비율이 가장 높은 변수명을 출력하시오. 

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv("data/3회/303_titanic.csv")
df.head()
```

### 방법 1. 

```python
rst_df = df.isnull().sum() / df.count()
rst_df.sort_values(ascending=False).index[0]
```

### 방법 2. 

```python
rst_df = df.isnull().sum().reset_index()
rst_df.columns = ['var', 'count']
rst_df
```

```python
result = rst_df.loc[rst_df['count'] == rst_df['count'].max(), 'var'].values[0]
result
```
